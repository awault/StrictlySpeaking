```python
def sinplot(n=10, flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, n + 1):
        plt.plot(x, np.sin(x + i * .5) * (n + 2 - i) * flip)
        plt.title('Sinewaves') # Chart Title
        #plt.figtext(x=0.7,y=0.15,s="Strictly Speaking", fontsize="small", ha='center', va='center', alpha=1, transform=plt.gcf().transFigure) # Watermark


print(sinplot())
```
